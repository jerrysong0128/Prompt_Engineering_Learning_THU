```python
import fitz  # PyMuPDF
import requests
from io import BytesIO

def extract_pdf_text(url: str, num_pages: int) -> str:
    # Download the PDF file
    response = requests.get(url)
    pdf_data = BytesIO(response.content)

    # Open the PDF file
    document = fitz.open(stream=pdf_data, filetype="pdf")

    # Extract text from the specified number of pages
    pages_text = []
    for page_number in range(num_pages):
        page = document.load_page(page_number)
        pages_text.append(page.get_text())

    # Combine the text from the specified pages
    combined_text = "\n".join(pages_text)

    return combined_text

# Example usage
url = "http://www.sse.com.cn/disclosure/listedinfo/announcement/c/new/2024-05-25/600885_20240525_X3DI.pdf"
num_pages = 1
extracted_text = extract_pdf_text(url, num_pages)
print(extracted_text)
```
