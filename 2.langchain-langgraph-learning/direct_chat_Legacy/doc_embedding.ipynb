```python
import requests

# Set the URL
url = 'http://47.253.94.63:8900/search_academic_db'

# Define the headers
headers = {
    'accept': 'application/json',
    'Authorization': 'Bearer <Authorization_key>',
    'Content-Type': 'application/json'
}

# Define the JSON payload
payload = {
    "query": "In this study, one tonne of cement and solid waste cement was used as a unit. Due to a large number of overlapping production links between the industrial products and other products, it was difficult to collect all input and output data, and there was no need to collect data having little impact on the results. Therefore, the boundaries of evaluation need to be defined. In this study, the system boundary was defined as “cradle to gate”,whichconsidered the process of the material mining, raw material preparation, clinker calcination, grinding and transportation, excluding cement use and waste stage. Solid waste initiated its life cycle from the abandonment stage. The system boundary is illustrated in Fig. 3.",
    "top_k": 10
}

# Send the POST request
response = requests.post(url, headers=headers, json=payload)

# Check if the request was successful
if response.status_code == 200:
    print("Request was successful!")
    # Print the response data
    print(response.json())
else:
    print("Request failed with status code:", response.status_code)
```
